```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('tripadvisor_review.csv')
print(df.head())
```

```python
df.columns = ['User ID', 'art galleries', 'dance clubs', 'juice bars', 'restaurants', 'museums', 'resorts', 'parks/picnic spots', 'beaches', 'theaters', 'religious institutions']
```

```python
print(df.head())
```

```python
sns.heatmap(df.corr())
```

```python
plt.scatter(df['User ID'].sample(n=20, random_state=1), df['juice bars'].sample(n=20, random_state=1))
plt.ylabel('juice bars')
plt.xlabel('User ID')
plt.xticks(rotation='vertical')
```

```python
plt.scatter(df['User ID'].sample(n=20, random_state=1), df['art galleries'].sample(n=20, random_state=1))
plt.ylabel('juice bars')
plt.xlabel('User ID')
plt.xticks(rotation='vertical')
```
